```python
import pandas as pd
import numpy as np

df = pd.DataFrame()

df['MANGERID'] = ['Sridhar','Rajkumar','Monish','Ranga','Hathija']
df['EMPID'] = ['Rajkumar','Monish','Hathija','Nish','Eniya']

final_df = pd.DataFrame(columns=['MANGERID','EMPID'])
level = 1
#for i in df['MANGERID']:
def fun(i,level):
    mask = np.column_stack([df[col].str.contains(i, na=False) for col in df])
    new_df = df.loc[mask.any(axis=1)]
    if len(new_df) == 1:
        return new_df['MANGERID'], level
    mang = ''.join(new_df[new_df['EMPID'] == i]['MANGERID'].values)
    level+=1
    return fun(mang,level)

    

def mngrEMP(i):
    mask = np.column_stack([df[col].str.contains(i, na=False) for col in df])
    new_df = df.loc[mask.any(axis=1)]
    if len(new_df) == 1:
        new_df['Level'] = 1
        global final_df
        final_df = final_df.append(new_df,ignore_index=True)
    else:
        new_df = df[df['MANGERID'] == i]
        mang = ''.join(df[df['EMPID'] == i]['MANGERID'].values)
        
        emp = ''.join(new_df[new_df['MANGERID'] == i]['EMPID'].values)
        manager, level = fun(mang,1)
        d = {'MANGERID':manager,
             'EMPID':emp,
             'Level':[level+1]}
        app_df = pd.DataFrame(d)
        final_df = final_df.append(app_df,ignore_index=True)
        
df['MANGERID'].apply(mngrEMP)
final_df
```
